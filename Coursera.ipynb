```python
duration_info = coursera_copy['course_time'].str.extract(r'(\d+)\s*-\s*(\d+)\s*(\D+)')
coursera_copy['duration_lower'] = duration_info[0].astype(float)
coursera_copy['duration_upper'] = duration_info[1].astype(float)
coursera_copy['duration_unit'] = duration_info[2]
conversion_rates = {'Months': 30, 'Weeks': 7}
coursera_copy['duration_in_days_lower'] = coursera_copy['duration_lower'] * coursera_copy['duration_unit'].map(conversion_rates).fillna(1)
coursera_copy['duration_in_days_upper'] = coursera_copy['duration_upper'] * coursera_copy['duration_unit'].map(conversion_rates).fillna(1)

# Distribute days proportionally between the lower and upper bounds
total_days = coursera_copy['duration_in_days_upper'] - coursera_copy['duration_in_days_lower']
coursera_copy['duration_in_days'] = coursera_copy['duration_in_days_lower'] + total_days * 0.5
print(coursera_copy[['course_time', 'duration_in_days']])
```

```python
sns.violinplot(x=coursera_copy['duration_in_days'])
plt.xlabel('Duration in Days')
plt.title('Distribution of Durations')
plt.show()
```
